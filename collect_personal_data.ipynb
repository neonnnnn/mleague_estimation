```python
import requests
from bs4 import BeautifulSoup
import json
import os
import datetime
import re
```

```python
url = 'https://m-league.jp/stats'
```

```python
save_dir = os.path.join('personal_data')
os.makedirs(save_dir, exist_ok=True)
```

```python
res = requests.get(url)
soup = BeautifulSoup(res.text, 'html.parser')
results_all = soup.find_all('div', {'class': 'p-stats__tableInner'})
```

```python
scores_all = {}
for result in results_all:
    names = {}
    scores = {}
    for tr in result.find_all("tr"):
        row_name = tr.find("th").get_text()
        if row_name == "選手名":
            print(row_name)
            for name in tr.find_all("th", {"scope": "col"}):
                names[len(names)] = name.get_text()
                scores_all[name.get_text()] = {}
        else:
            for i, val in enumerate(tr.find_all("td")):
                scores_all[names[i]][row_name] = val.get_text()
print(scores_all)
```
